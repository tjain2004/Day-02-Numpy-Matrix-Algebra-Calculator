```python
#### import numpy as np

# -----------------------------------------
# 1. Take Matrix Input from User
# -----------------------------------------
def input_matrix():
    rows = int(input("Enter number of rows: "))
    cols = int(input("Enter number of columns: "))
    print("Enter matrix values row by row (space separated):")

    matrix = []
    for _ in range(rows):
        row = list(map(float, input().split()))
        matrix.append(row)

    return np.array(matrix)

# -----------------------------------------
# 2. Basic Matrix Operations
# -----------------------------------------

def matrix_addition():
    print("Enter Matrix A:")
    A = input_matrix()
    print("Enter Matrix B:")
    B = input_matrix()
    print("A + B =\n", A + B)

def matrix_subtraction():
    print("Enter Matrix A:")
    A = input_matrix()
    print("Enter Matrix B:")
    B = input_matrix()
    print("A - B =\n", A - B)

def matrix_multiplication():
    print("Enter Matrix A:")
    A = input_matrix()
    print("Enter Matrix B:")
    B = input_matrix()
    print("A × B =\n", A @ B)

def scalar_multiplication():
    print("Enter Matrix A:")
    A = input_matrix()
    k = float(input("Enter scalar value: "))
    print("k × A =\n", k * A)

# -----------------------------------------
# 3. Advanced Operations
# -----------------------------------------

def matrix_transpose():
    print("Enter Matrix:")
    A = input_matrix()
    print("Transpose =\n", A.T)

def matrix_determinant():
    print("Enter Square Matrix:")
    A = input_matrix()
    print("Determinant =", np.linalg.det(A))

def matrix_rank():
    print("Enter Matrix:")
    A = input_matrix()
    print("Rank =", np.linalg.matrix_rank(A))

def matrix_inverse():
    print("Enter Square Matrix:")
    A = input_matrix()
    try:
        print("Inverse =\n", np.linalg.inv(A))
    except:
        print("Matrix is singular — inverse does NOT exist.")

def eigen_values_vectors():
    print("Enter Square Matrix:")
    A = input_matrix()
    values, vectors = np.linalg.eig(A)
    print("Eigenvalues:\n", values)
    print("Eigenvectors:\n", vectors)

# -----------------------------------------
# 4. Solve System of Linear Equations
# -----------------------------------------
def solve_linear_equations():
    print("Enter Coefficient Matrix A:")
    A = input_matrix()
    print("Enter Constant Matrix b (column vector):")
    b = input_matrix()
    
    try:
        x = np.linalg.solve(A, b)
        print("Solution x =\n", x)
    except:
        print("No unique solution exists for Ax = b")

# -----------------------------------------
# 5. MAIN MENU
# -----------------------------------------
def main():
    while True:
        print("\nMatrix Algebra System (MAS)")
        print("----------------------------")
        print("1. Matrix Addition")
        print("2. Matrix Subtraction")
        print("3. Matrix Multiplication")
        print("4. Scalar Multiplication")
        print("5. Transpose")
        print("6. Determinant")
        print("7. Inverse")
        print("8. Rank")
        print("9. Eigenvalues & Eigenvectors")
        print("10. Solve System of Equations (Ax = b)")
        print("11. Exit")

        choice = int(input("Choose an option: "))

        if choice == 1: matrix_addition()
        elif choice == 2: matrix_subtraction()
        elif choice == 3: matrix_multiplication()
        elif choice == 4: scalar_multiplication()
        elif choice == 5: matrix_transpose()
        elif choice == 6: matrix_determinant()
        elif choice == 7: matrix_inverse()
        elif choice == 8: matrix_rank()
        elif choice == 9: eigen_values_vectors()
        elif choice == 10: solve_linear_equations()
        elif choice == 11:
            print("Exiting...")
            break
        else:
            print("Invalid choice. Try again.")

main()
```
